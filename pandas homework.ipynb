```python
def read(df,item,i):
    t=pd.read_csv(df[df["Title"]==item]["CSV"][i])
    print("Info")
    print(t.info())
    print("\n\nDescribe")
    print(t.describe())
    print("\n\nObject Describe")
    try:
        print(t[t.dtypes[t.dtypes=="object"].index].describe())
    except:
        print("Not found")
```
